# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def ReLU(x):  # 這邊的X，可放array，若用max(0, x) 則不行
    return abs(x) * (x > 0) 

def dReLU(x):
    return (1 * (x > 0))

plt.figure(figsize=(10, 10))

x = plt.linspace(-10,10,100)

plt.plot(x, ReLU(x), 'b', label='ReLU')
plt.plot(x, dReLU(x), 'r', label='dReLU')

plt.grid()

plt.title('ReLU Function')

plt.text(0, 9, r'$f(x)= (abs(x) * (x > 0))$', fontsize=20)
plt.text(2, 1.6, r"$f'(x)= (1 * (x > 0))$", fontsize=20)

plt.show()
```
